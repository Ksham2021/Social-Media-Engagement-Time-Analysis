```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("social_media_dummy_data.csv")
```

```python
df["enagagement"] = df["comments"] + df["likes"] + df["shares"]
```

```python
hourly_enagagement = df.groupby("post_time")["enagagement"].mean()
```

```python
plt.plot(hourly_enagagement.index, hourly_enagagement.values)
plt.xlabel("Posting Time (Hour)")
plt.ylabel("Average Engagement")
plt.title("Engagement vs Posting Time")
plt.show()
```

```python
best_time = hourly_enagagement.idxmax()
```

```python
print("Best time to post: ",best_time)
```
